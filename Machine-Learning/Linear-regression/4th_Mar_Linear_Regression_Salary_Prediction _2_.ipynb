```python
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
```

```python
data = pd.read_csv("Salary_Data _1_.csv")
data.head()
```

```python
data.describe(percentiles=[.1, .2, .5, .8])
```

```python
print(data.isnull().sum())
data.shape
```

### Split into inputs(x) and output(y)

```python
# x = data[['YearsExperience']].values
# x.shape
# type(x)

x = data[['YearsExperience']]
x.shape
```

```python
y = data[['Salary']]
y.shape
```

```python
## Split into training and testing (80:20)
# from sklearn.model_selection import train_test_split

xtrain,xtest,ytrain,ytest = sklearn.model_selection.train_test_split(x,y,train_size=0.8,random_state=0)
```

```python
xtest
```

```python
ytest
```

```python
# ### Make the model
# from sklearn.linear_model import LinearRegression
# model = LinearRegression()

# ## Train the model
# model.fit(xtrain,ytrain)

model = sklearn.linear_model.LinearRegression()

## Train the model
model.fit(xtrain,ytrain)
```

```python
## Test the model (Prediction)
ypred = model.predict(xtest)
ypred
```

```python
ytest
```

```python
## Evaluation
# from sklearn.metrics import r2_score

r2 = sklearn.metrics.r2_score(ytest,ypred)
# r2 = sklearn.metrics.mean_squared_error(ytest,ypred)
r2
```

```python
## Line of regression (training samples)
plt.scatter(xtrain,ytrain,color='red',label='Actual')
plt.scatter(xtrain,model.predict(xtrain),color='blue',label='Prediction')
plt.plot(xtrain,model.predict(xtrain),color='orange',label='LOR',linestyle='dotted')
plt.legend()
plt.show()
```
